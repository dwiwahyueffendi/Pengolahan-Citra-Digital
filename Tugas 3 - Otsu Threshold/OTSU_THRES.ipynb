```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2

img = cv2.imread("org.jpg", 0)
x, res = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.imshow(res, 'gray')
```

```python
#CONVERT GRAYSCALE TO BINER
imgGray = cv2.imread('kucingGray.jpg', 0)
ret, biner = cv2.threshold(imgGray,30,255,cv2.THRESH_BINARY)
ret, otsu = cv2.threshold(imgGray,0,255,cv2.THRESH_OTSU)

#SAVE HASIL GREYSCALE
cv2.imwrite("kucingBiner.jpg", biner)
cv2.imwrite("kucingOtsu.jpg", otsu)

# menampilkan hasil
titles = ['IMAGE GRAYSCALE', 'BINER', 'OTSU']
images = [imgGray, biner, otsu]

# menampilkan beberapa gambar sekaligus
for i in range(3):
    # 3 baris, 2 kolom
    plt.subplot(2,3,i+1),
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
